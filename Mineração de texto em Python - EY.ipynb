```python
#bibliotecas necessárias 

import matplotlib.pyplot as plt
import nltk
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
import re

## Obs. Caso não tenha biblioteca instalado, poderá utilizar 
### !pip install (nome da biblioteca) 
#### no proprio Jupyter Notebook
```

```python
#Criando nossa variável Corpus que servirá como base para nossa análise 
corpus = PlaintextCorpusReader('Relatorio', '.*', encoding = "ISO-8859-1")
```

```python
#visualizando o nome do nosso arquivo
arquivos = corpus.fileids()

#Abaixo demonstra quantos nomes de arquivos queremos ver. 
arquivos[0:1]

##Como só temos um arquivo, não é necessário colocar mais que 1
```

```python
#Visualizando o texto que contém em determinado arquivo
texto1 = corpus.raw('relatorio-anual-2020-EY.txt')
texto1

## Devido ser um arquivo muito grande(com muitas palavras),
### fica inviável a visualização do texto
```

```python
#Acessando todas as palavras de todos os arquivos 
todas_palavras = corpus.raw()
```

```python
#Obtendo todas as palavras dos arquivos
palavras = corpus.words()

#Verificando qual palavra está no indice 9
palavras[1501]
```

```python
#Vendo o total de quantidade de palavras
len(palavras)
```

```python
#Verificando as stopwords em português e criando uma variável para ela
stop_words = stopwords.words('portuguese')
stop_words[0:10]
```

```python
mapa_cores = ListedColormap(['yellow', 'white'])
nuvem_palavras = WordCloud(background_color = 'black',
                          colormap = mapa_cores,
                          stopwords = stop_words,
                          max_words = 190)
nuvem_palavras.generate(todas_palavras)
plt.imshow(nuvem_palavras)
```
